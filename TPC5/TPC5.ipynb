```python
import matplotlib.pyplot as plt


def calcVar():

    file = open("myheart.csv", "r")
    file.readline()
    texto = file.read().split("\n")

    male = 0
    female = 0
    maleSick = 0
    femaleSick = 0
    lowerAge = 100
    upperAge = 0
    lowerCol = 300
    upperCol = 0

    for linha in texto:
        line = linha.split(",")
        if int(line[0]) < lowerAge:     #Valores mínimos e máximos de idade
            lowerAge = int(line[0])
        if int(line[0]) > upperAge:
            upperAge = int(line[0])
        if line[1] == "M":              #Total sexos
            male = male +1
        if line[1] == "F":
            female = female +1
        if int(line[3]) != 0:           
            if int(line[3]) < lowerCol:     #Valore mínimos e máximos de colesterol
                lowerCol = int(line[3])
            if int(line[3]) > upperCol:
                upperCol = int(line[3])
            if line[1] == "M":              #Doentes por sexos
                maleSick = maleSick +1
            if line[1] == "F":
                femaleSick = femaleSick +1
    
    res = (male, female, maleSick, femaleSick, lowerAge, upperAge, lowerCol, upperCol)
    file.close()
    return res

def valoresGraph(vars):

    file = open("myheart.csv", "r")
    file.readline()
    texto = file.read().split("\n")

    male, female, maleSick, femaleSick, lowerAge, upperAge, lowerCol, upperCol = vars

    while lowerCol % 10 !=0:                #Ajuste da escala de colesterol
        lowerCol = lowerCol -1
    while upperCol % 10 !=0:
        upperCol = upperCol +1

    while lowerAge % 5 != 0:                #Ajuste da escala de idade
        lowerAge = lowerAge -1
    while upperAge % 5 != 0:
        upperAge = upperAge +1

    col = [x for x in range(lowerCol, upperCol +10, 10)]    #Intervalo de idades
    idades = [x for x in range(lowerAge, upperAge +5, 5)]   #Intervalo de colesterol
    distribIdades = []
    distribCol = []

    for i in idades:        #Num de pessoas doentes por intervalo de idades
        idadeSick = 0
        for linha in texto:
            line = linha.split(",")
            if int(line[0]) in range(i, i +5):
                if int(line[5]) == 1:
                    idadeSick = idadeSick +1
        distribIdades.append(idadeSick)

    for i in col:           #Num de pessoas doentes por intervalo de colesterol
        colSick = 0
        for linha in texto:
            line = linha.split(",")
            if int(line[3]) in range(i, i +10):
                if int(line[5]) == 1:
                    colSick = colSick +1
        distribCol.append(colSick)

    colGraph = ["["+str(x)+","+str(x+9)+"]" for x in col]           #Criar eixo xx
    idadesGraph = ["["+str(x)+","+str(x+4)+"]" for x in idades]     #Criar eixo yy

    res = (idadesGraph, distribIdades, colGraph, distribCol)
    file.close()
    return res

#Programa para imprimir os graficos

function=int(input("O que pretende fazer?"))
print('''Menu:
(1) Gráfico de distribuição de pessoas doentes por idade.
(2) Gráfico de distribuição de pessoas doentes por nível de colesterol.
(0) Sair''')

while function != 0:
    idadesGraph, distribIdades, colGraph, distribCol = valoresGraph(calcVar())
    if function == 1:
        fig1 = plt.figure(figsize = (10, 5))
        plt.bar(idadesGraph, distribIdades, color = "red", width = 0.5)
        plt.xlabel("Idade")
        plt.ylabel("Num de pessoas doentes")
        plt.title("Número de pessoas doentes em função da idade")
        plt.show()
    if function == 2:
        fig2 = plt.figure(figsize = (50, 25))
        plt.bar(colGraph, distribCol, color = "blue", width = 0.5)
        plt.xlabel("Nível de colesterol")
        plt.ylabel("Num de pessoas doentes")
        plt.title("Número de pessoas doentes em função dos níveis de colesterol")
        plt.show()
    function=int(input("O que pretende fazer?"))
```
