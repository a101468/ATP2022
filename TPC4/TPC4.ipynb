```python
def menu():
    print("""Bem-vindo aos cinemas ATP2022!
Menu:
(1) Criar novo cinema
(2) Criar nova sala
(3) Procurar filme
(4) Ver filmes em exibição
(5) Verificar disponibilidade da sala
(6) Comprar bilhete
(7) Lista de cinemas
(8) Remover cinema
(9) Remover sala
(0) Sair""")
    return

def criarCinema(nomeCinema):
    cinemas[nomeCinema]={}
    return

def criarSala(nomeCinema,nomeSala,lugares,nomeFilme):
    newSala=(lugares,[],nomeFilme)
    cinemas[nomeCinema][nomeSala]=newSala
    return

def venderBilhete(nomeCinema,nomeFilme,lugar):
    for sala in cinemas[nomeCinema]:
        lugares,vendidos,filme=cinemas[nomeCinema][sala]
        if nomeFilme==filme:
            vendidos.append(lugar)
    return

def filmesCinema(nomeCinema):
    filmes=[]
    for sala in cinemas[nomeCinema]:
        lugares,vendidos,nomeFilme=cinemas[cinema][sala]
        filmes.append(nomeFilme)
    return filmes

def procurarFilme(nomeFilme):
    res=[]
    for cinema in cinemas:
        for sala in cinemas[cinema]:
            lugares,vendidos,filme=cinemas[cinema][sala]
            if filme==nomeFilme:
                res.append((cinema,sala,lugares-len(vendidos)))
    if len(res)==0:
        return False
    return res

def listarDisponibilidades(nomeCinema,nomeFilme):
    for sala in cinemas[nomeCinema]:
        lugares,vendidos,filme=cinemas[nomeCinema][sala]
        if filme==nomeFilme:
            res=(lugares-len(vendidos),vendidos)
    return res

def listarCinema():
    res=[]
    for item in cinemas.items():
        res.append(item[0])
    return res

def removerCinema(nomeCinema):
    cinemas.pop(nomeCinema)
    return

def removerSala(nomeCinema,nomeSala):
    cinemas[nomeCinema].pop(nomeSala)
    return

menu()
cinemas={}
function=int(input("O que pretende fazer?"))

while function!=0:
    if function==1:
        nomeCinema=input("Qual o nome do novo cinema?")
        if nomeCinema in cinemas:
            print("\nJá existe um cinema com esse nome.")
        else:
            criarCinema(nomeCinema)
    if function==2:
        nomeCinema=input("Em que cinema pretende criar uma nova sala?")
        if nomeCinema not in cinemas:
            print("\nO cinema que procura não existe.")
        else:
            nomeSala=input("Qual o nome da nova sala?")
            if nomeSala in cinemas[nomeCinema]:
                print("\nJá existe uma sala com esse nome.")
            else:
                criarSala(nomeCinema,nomeSala,int(input("Quantos lugares terá a nova sala?")),input("Que filme será exibido na nova sala?"))
    if function==3:
        filme=input("Qual o filme que procura?")
        if procurarFilme(filme)!=False:
            for item in procurarFilme(filme):
                cinema,sala,lugaresFree=item
                print("\nO filme",filme,"está em exibição na sala",sala,"e tem",lugaresFree,"lugares disponíveis.")
        else:
            print("\nO filme que procura não existe.")
    if function==4:
        nomeCinema=input("De que cinema pretende ver os filmes em exibição?")
        if nomeCinema in cinemas:
            print("\nNo cinema",nomeCinema,"estão em exibição os seguintes filmes:","\n",filmesCinema(nomeCinema))
        else:
            print("\nO cinema que pretende ver não existe.")
    if function==5:
        nomeCinema=input("Qual o cinema que procura?")
        if nomeCinema in cinemas:
            nomeFilme=input("De que filme procura ver os lugares disponíveis?")
            if nomeFilme in filmesCinema(nomeCinema):
                print("\nO filme",nomeFilme,"tem",listarDisponibilidades(nomeCinema,nomeFilme)[0],"lugares disponíveis, a lista de lugares ocupados é:\n",listarDisponibilidades(nomeCinema,nomeFilme)[1])
            else:
                print("\nO filme que procura não está em exibição neste cinema.")
        else:
            print("\nO cinema que procura não existe.")
    if function==6:
        nomeCinema=input("Para que cinema pretende comprar bilhete?")
        if nomeCinema in cinemas:
            nomeFilme=input("Para que filme pretende comprar bilhete?")
            for sala in cinemas[nomeCinema]:
                lugares,vendidos,filme=cinemas[nomeCinema][sala]
                if filme==nomeFilme:
                    lugar=int(input("Para que lugar pretende comprar bilhete?"))
                    if lugar in range(lugares):
                        venderBilhete(nomeCinema,nomeFilme,lugar)
                    else:
                        print("\nO lugar que pretende comprar não existe.")
                else:
                    print("\nO filme para o qual pretende comprar bilhete não está em exibição nesse cinema.")
        else:
            print("\nO cinema que procura não existe.")
    if function==7:
        print("\nA lista de cinemas é:\n",listarCinema())
    if function==8:
        nomeCinema=input("Qual o cinema que pretende remover?")
        if nomeCinema in cinemas:
            removerCinema(nomeCinema)
        else:
            print("\n","O cinema que pretende remover não existe.")
    if function==9:
        nomeCinema=input("Qual o cinema de onde pretende remover uma sala?")
        if nomeCinema in cinemas:
            nomeSala=input("Qual a sala que prentede remover?")
            if nomeSala in cinemas[nomeCinema]:
                removerSala(nomeCinema,nomeSala)
            else:
                print("\nA sala que pretende remover não existe.")
        else:
            print("\nO cinema que procura não existe.")
    function=int(input("O que pretende fazer?"))
print("\nObrigado por procurar os nossos cinemas!\nAté à próxima!")
```
